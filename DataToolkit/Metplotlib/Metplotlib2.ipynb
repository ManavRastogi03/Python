```python
#area plot >> variable chages over time 
years=[2021,2022,2017,2012]
category1=[20,25,30,10]
category2=[2,25,20,11]
category3=[10,5,25,5]
plt.stackplot(years,category1,category2,category3 ,labels=["category1","category2","category3"],
              colors=['skyblue','lightgreen','orange'])
plt.xlabel("years")

plt.ylabel("Values")
plt.title("Area Plot")
plt.legend()
plt.show()
```
