```python
#stacked histogram >>two histogram together 
data1=np.random.normal(0,1,5000)
data2=np.random.normal(3,2,5000)
```

```python
plt.hist([data1,data2],color=['r','b'],bins=20,label=['dist 1','dist 2'])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid()
plt.title('stacked histogram')
```

```python
#bar plot >> to see the distribution of the categorical variable
x=['a','b','c','d'] #name of the student
y=[4,2,3,7] #no of days present
plt.bar(x,y,color='r')#vertical plot
```

```python
#pie chart is used to under stand the distribution of the data
lngs=["C","Pyhton","Java"]
stu=[20,100,40]
plt.pie(stu,labels=lngs,autopct="%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
#explore on one segment 
explode=(0.0,0.1,0.1)
lngs=["C","Pyhton","Java"]
stu=[20,100,40]
plt.pie(stu,labels=lngs,autopct="%1.1f%%",explode=explode)
plt.title("Pie chart")
plt.show()
```
