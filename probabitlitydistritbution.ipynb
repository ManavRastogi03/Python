```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#normal Distribution>> gaussion distribution>> majpity of the data point is cluster around mean
mu=0
sigma=1
l=np.random.normal(mu,sigma,size=10000)
```

```python
len(l)
```

```python
plt.hist(l,30)
plt.show()
```

```python
sns.displot(l)
```

```python
#uniform distribution>> all the outcome have equal prob to occur in a range
y=np.random.uniform(0,1,10000)
plt.hist(y)
plt.show()
```

```python
sns.displot(y)
```

```python
#log noraml distribution
y=np.random.lognormal(mean=0,sigma=0.5,size=1000)
```

```python
plt.hist(y,100)
plt.show()
```

```python
sns.boxplot(y) #since log normal is right skwed so outlier are present
```

```python
sns.displot(y)
```

```python
sns.boxplot(y) #since lognormal is right skwed, so the outlier are present 
```

```python
#lognormal >> normal distribution
sns.displot(np.log(y))
```

```python
#square root transformation
sns.displot(np.sqrt(y))
```

```python
#chi-square distribution
c=np.random.chisquare(df=30,size=10000)
d=np.random.chisquare(df=10,size=10000)
c=c[np.where(abs(c)<100)]
d=d[np.where(abs(d)<100)]
plt.hist(c,1000)
plt.hist(d,1000)
plt.show()
```

```python
#t distribution
t=np.random.standard_t(df=3,size=1000)
t=t[np.where(abs(t)<100)]
plt.hist(t,100)
plt.show()
```

```python
#F distribution
f=np.random.f(dfnum=100,dfden=20,size=1000)
f=f[np.where(abs(f)<10)]
plt.hist(f,1000)
plt.show()
```

```python
#discrete distribution
#bernollii distribution>>binomial dist with one trial is bernoulli distributioon
b=np.random.binomial(n=1,p=0.5,size=1000)
plt.hist(b)
plt.show()
```

```python
#binomial distribution
b=np.random.binomial(n=100,p=0.5,size=1000)
plt.hist(b)
plt.show()
```

```python
#possions distribution
p=np.random.poisson(lam=5,size=1000)
plt.hist(p,20)
plt.show()
```
