```python
cat_features=X.select_dtypes(include='object').columns
```

```python
nums_features=X.select_dtypes(exclude='object').columns
```

```python
nums_features
```

```python
#scaling>> nnumerical features
#categorical>>Onehot encoding
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
#OnehotEncoding it tranform the string into numerical 
#StandardScaler it tranfor the one scale to another scale
# colTransformer applies the tranformation to coloumn of an array or pandas data frame

```

```python
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(drop="first")
preprocessor=ColumnTransformer(
    [
        ('OneHotEncoder',oh_transformer,cat_features),
        ('StandardScaler',numeric_transformer,nums_features)
    ]
)
```

```python
preprocessor
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.transform(X_test)
```
