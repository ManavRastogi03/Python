```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6])
y=np.array([1,3,5,7,9,10])

plt.scatter(x,y)
```

```python
#linear interpolation >>linear in nature
x_new=np.linspace(1,5,10) #create a new x value
y_inter=np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_inter)
```

```python
#cubic interpolation
x=np.array([1,2,3,4,5,6])
y=np.array([1,8,27,64,125,216])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind="cubic")
```

```python
f
```

```python
x_new=np.linspace(1,5,10)
y_interpl=f(x_new)
```

```python
plt.scatter(x_new,y_inter)
```

```python
x_new=np.linspace(x,y,2)
y_interp=np.polyval(p,x_new)
```
