```python
data=pd.DataFrame(daibetes.data,columns=daibetes.feature_names)
```

```python
data["target"]=daibetes.target
```

```python
#EDA, DATACLEANINF,DATAPREPRATION,FEATURE ENGINEERING
```

```python
X=data['bmi'] #one indepedent variable
y=data["target"] #one depedent
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)

#random state when fixed help us to reproduce the sae result again and again
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# assuming X_train is a single column Series
X_train = X_train.values.reshape(-1, 1)  # or pd.DataFrame(X_train)
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python

y_pred=model.predict(X_test)
```

```python
#visualize the prediction
plt.scatter(X_test,y_test,color="red",label="Actual")
plt.plot(X_test,y_pred,color="blue",linewidth=2,label="Predicted")
plt.xlabel("BMI")
plt.ylabel("One year progerssion target")
plt.title("Linear regission on dibetes data")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
